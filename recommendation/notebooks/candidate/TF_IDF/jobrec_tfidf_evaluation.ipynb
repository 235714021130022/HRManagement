```python

# ===== Step 8: Evaluation metrics @K (baseline vs improved) =====
# topK job: bao nhiêu job đúng ngành so với CV
def _taxonomy_hit_at_k(df: pd.DataFrame) -> float:
    return float((df["job_tax"] == df["cv_tax"]).mean())

# topK job: bao nhiêu job lệch level so với CV 
def _seniority_mismatch_at_k(df: pd.DataFrame) -> float:
    return float((df["job_sen"] != df["cv_sen"]).mean())

# topK job: trùng bao nhiêu skill với CV
def _avg_skill_overlap_at_k(df: pd.DataFrame, cv_index: int) -> float:
    cv_sk = extract_skills(cv_texts_norm.iloc[cv_index])
    if not cv_sk:
        return 0.0
    overlaps = df["job_skills"].apply(lambda s: len(s & cv_sk) / max(1, len(cv_sk)))
    return float(overlaps.mean())

# matches có tồn tại không và bao nhiêu? 
def _explainability_rate_at_k(df: pd.DataFrame) -> float:
    return float(df["matched_skills"].apply(lambda x: isinstance(x, list) and len(x) > 0).mean())

def enrich_for_eval(df: pd.DataFrame, cv_index: int) -> pd.DataFrame:
    """Add job_tax/job_sen/job_skills + cv_tax/cv_sen + matched/missing for evaluation metrics."""
    out = df.copy()

    # ensure index maps to jobs row indices (this is true if df came from jobs.iloc[idx].copy())
    out["title"] = out["title"].fillna("").astype(str)

    # If cleaned_description not already in df, pull from global jobs by index
    if "cleaned_description" not in out.columns:
        out["cleaned_description"] = jobs.loc[out.index, "cleaned_description"].fillna("").astype(str)

    job_text_series = out["title"] + " " + out["cleaned_description"]

    # CV labels
    cv_text = cv_texts_norm.iloc[cv_index]
    out["cv_tax"] = cv_taxonomy(cv_index)
    out["cv_sen"] = infer_seniority_from_cv(cv_text)
    cv_sk = extract_skills(cv_text)

    # Job labels
    out["job_tax"] = out.apply(lambda r: classify_taxonomy_title_desc(r["title"], r["cleaned_description"]), axis=1)
    out["job_sen"] = out["title"].apply(infer_seniority_from_title)
    out["job_skills"] = job_text_series.apply(extract_skills)

    # Explain
    out["matched_skills"] = out["job_skills"].apply(lambda s: sorted(list(s & cv_sk))[:5] if isinstance(s, set) else [])
    out["missing_skills"] = out["job_skills"].apply(lambda s: sorted(list(s - cv_sk))[:5] if isinstance(s, set) else [])

    return out
 
def evaluate_option2_tfidf(sample_size: int = 50, top_k: int = 10, seed: int = 42):
    rng = np.random.default_rng(seed)
    n = len(resumes)
    idxs = rng.choice(n, size=min(sample_size, n), replace=False)

    rows = []
    for cv_idx in idxs:
        base = baseline_topk_tfidf(cv_idx, top_k=top_k)
        imp  = recommend_jobs_option2_tfidf(cv_idx, top_k=top_k)

        base2 = enrich_for_eval(base, cv_idx)
        imp2  = enrich_for_eval(imp,  cv_idx)

        rows.append({
            "type": "baseline",
            "cv_index": int(cv_idx),
            "TaxonomyHit@K": _taxonomy_hit_at_k(base2),
            "SeniorityMismatch@K": _seniority_mismatch_at_k(base2),
            "AvgSkillOverlap@K": _avg_skill_overlap_at_k(base2, cv_idx),
            "ExplainabilityRate@K": _explainability_rate_at_k(base2),
        })
        rows.append({
            "type": "improved",
            "cv_index": int(cv_idx),
            "TaxonomyHit@K": _taxonomy_hit_at_k(imp2),
            "SeniorityMismatch@K": _seniority_mismatch_at_k(imp2),
            "AvgSkillOverlap@K": _avg_skill_overlap_at_k(imp2, cv_idx),
            "ExplainabilityRate@K": _explainability_rate_at_k(imp2),
        })


    metrics_df = pd.DataFrame(rows)
    summary_df = metrics_df.groupby("type")[["TaxonomyHit@K","SeniorityMismatch@K","AvgSkillOverlap@K","ExplainabilityRate@K"]].mean()
    return metrics_df, summary_df

metrics_df, summary_df = evaluate_option2_tfidf(sample_size=50)
summary_df
```

### Diễn giải TaxonomyHit@K
- Tỷ lệ các công việc trong top-K thuộc đúng ngành/ngữ nghĩa nghề nghiệp của hồ sơ ứng viên.

| Giá trị TaxonomyHit@K | Đánh giá |
|----------------------|---------|
| < 0.40 | Rất kém – Phần lớn job gợi ý lệch ngành |
| 0.40 – 0.60 | Mức chấp nhận được (baseline) |
| 0.60 – 0.75 | Tốt – Đa số job đúng ngành |
| **0.75 – 0.90** | **Rất tốt – Tính nhất quán ngành nghề cao** |
| > 0.90 | Xuất sắc – Gần như hoàn toàn đúng ngành |

- TopK(cv) = danh sách K job được gợi ý cho CV

- tax(job) = taxonomy của job

- tax(cv) = taxonomy của CV

TaxonomyHit@K
= (1 / K) * Σ I( tax(job_i) == tax(cv) )


### Diễn giải SeniorityMismatch@K (càng thấp càng tốt)

- Tỷ lệ các công việc trong top-K không phù hợp cấp độ kinh nghiệm so với hồ sơ ứng viên.

| Giá trị SeniorityMismatch@K | Đánh giá |
|----------------------------|---------|
| > 0.90 | Rất kém – Lệch cấp độ nghiêm trọng |
| 0.75 – 0.90 | Kém – Kiểm soát seniority chưa tốt |
| 0.60 – 0.75 | Trung bình – Vẫn còn lệch cấp độ |
| **0.40 – 0.60** | **Tốt – Phù hợp cấp độ tương đối** |
| < 0.40 | Rất tốt – Khớp cấp độ cao |

SeniorityMismatch@K
= (1 / K) * Σ I( sen(job_i) ≠ sen(cv) )


### Diễn giải AvgSkillOverlap@K

- Tỷ lệ trung bình các kỹ năng của ứng viên xuất hiện trong mô tả công việc.

| Giá trị AvgSkillOverlap@K | Đánh giá |
|--------------------------|---------|
| < 0.40 | Khớp kỹ năng yếu |
| 0.40 – 0.55 | Khớp kỹ năng hạn chế |
| 0.55 – 0.70 | Khớp kỹ năng ở mức hợp lý |
| **0.70 – 0.85** | **Khớp kỹ năng tốt** |
| > 0.85 | Khớp kỹ năng rất tốt |

- Skills(cv) = tập kỹ năng CV

- Skills(job) = tập kỹ năng job

SkillOverlap(job, cv)
= |Skills(job) ∩ Skills(cv)| / |Skills(cv)|

AvgSkillOverlap@K
= (1 / K) * Σ SkillOverlap(job_i, cv)


### Diễn giải ExplainabilityRate@K

- Tỷ lệ các công việc gợi ý có thể giải thích được lý do gợi ý (dựa trên taxonomy, seniority, kỹ năng).

| Giá trị ExplainabilityRate@K | Đánh giá |
|-----------------------------|---------|
| < 0.80 | Khả năng giải thích thấp |
| 0.80 – 0.90 | Giải thích một phần |
| 0.90 – 0.98 | Giải thích tốt |
| **≥ 0.98** | **Giải thích đầy đủ** |
